```python
import matplotlib.pyplot as plt
import numpy as np

# Personality traits 
traits = ['Agreeableness', 'Conscientiousness', 'Extraversion', 'Neuroticism', 'Openness']

# F1 scores from both models
f1_tfidf = [0.85, 0.72, 0.90, 0.88, 0.79]  # TF-IDF model
f1_bert  = [0.82, 0.65, 0.86, 0.89, 0.77]  # BERT model

x = np.arange(len(traits)) 
width = 0.35  # width of the bars
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, f1_tfidf, width, label='TF-IDF + LR', color='skyblue')
bars2 = ax.bar(x + width/2, f1_bert, width, label='BERT', color='orange')

# Labels and aesthetics
ax.set_ylabel('F1 Score')
ax.set_title('F1 Score Comparison by Personality Trait')
ax.set_xticks(x)
ax.set_xticklabels(traits)
ax.set_ylim(0, 1.0)
ax.legend()

# Annotate bars with values
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
